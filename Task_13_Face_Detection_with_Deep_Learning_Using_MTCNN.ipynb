## **Face Detection With Deep Learning**
A number of deep learning methods have been developed and demonstrated for face detection.

Perhaps one of the more popular approaches is called the “Multi-Task Cascaded Convolutional Neural Network,” or MTCNN for short, described by Kaipeng Zhang, et al. in the 2016 paper titled “Joint Face Detection and Alignment Using Multitask Cascaded Convolutional Networks.”

The MTCNN is popular because it achieved then state-of-the-art results on a range of benchmark datasets, and because it is capable of also recognizing other facial features such as eyes and mouth, called landmark detection.

The network uses a cascade structure with three networks; first the image is rescaled to a range of different sizes (called an image pyramid), then the first model (Proposal Network or P-Net) proposes candidate facial regions, the second model (Refine Network or R-Net) filters the bounding boxes, and the third model (Output Network or O-Net) proposes facial landmarks.

```python

# face detection with mtcnn on a photograph
from matplotlib import pyplot
from mtcnn.mtcnn import MTCNN
# load image from file
filename = '/content/Selena gomez.jpg'
pixels = pyplot.imread(filename) # reading the image file
# create the detector, using default weights
detector = MTCNN()  #here we are instantiating MTCNN, We can pass our model file in that with the parameter "weights_file"(weights_file='modelfile.npy'), By, default, It has its own weights file, Here we didnt give any  model file,So it takes its default model weights_file
# detect faces in the image
faces = detector.detect_faces(pixels)  # It results the dictionary of values,it has the following values 
print(faces)                                      #1.Face box values
                                                  #2.confidence-The probability confidence of the prediction.
                                                  #3.keypoints-  Providing a dict with dots for the ‘left_eye‘, ‘right_eye‘, ‘nose‘, ‘mouth_left‘, and ‘mouth_right‘.
```

```python
# for viewing multi detected faces in a single image--they all stored in form of list of dictionaries.  ---Each dictionary contains info of one detected thing.
for face in faces:
  print(face)
```

### **Simple Code for Detect Face with rectangle Box and some Face points**

Since in the face dictionary,in 'keypoints' dictionary, it contains the predicted  co-ordinates of left eye,right eye,left mouth,right mouth,nose in the form of key value pair

Testing with single face image

```python

from matplotlib import pyplot
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from mtcnn.mtcnn import MTCNN


filename = '/content/Selena gomez.jpg'
img = pyplot.imread(filename)

detector = MTCNN()
faces = detector.detect_faces(img)

 
pyplot.imshow(img)
	
ax = pyplot.gca()  # get the context for drawing boxes
	
for face in faces:
	# get coordinates
	x, y, width, height = face['box']
		# create rectangle (Face box)
	rect = Rectangle((x, y), width, height, fill=False, color='blue')
	# draw the facebox(rectangle)
	ax.add_patch(rect)
		# draw the Circle(dots) at left eye,right eye,nose,left mouth ,right mouth
	for key, value in face['keypoints'].items():
		# create and draw dot for each point
		dot = Circle(value, radius=6, color='orange')
		ax.add_patch(dot)
	# show the plot
pyplot.show()
 
```

### **Extract detected face in seperate images**

Testing with Single face images

```python
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from mtcnn.mtcnn import MTCNN
 
# draw each face separately


filename = '/content/Selena gomez.jpg'
# load image from file
img = pyplot.imread(filename)
# create the detector, using default weights
detector = MTCNN()
# detect faces in the image
faces = detector.detect_faces(img)


# plot each face as a subplot
for i in range(len(faces)):
	# get coordinates
	x1, y1, width, height = faces[i]['box']
	x2, y2 = x1 + width, y1 + height           #with the x1 and y1 co-ordinate of corner point ,we can get x2 and y2 point,...By adding x1 with width gives x2 and y1 with height gives y2,With two corner point we can get rectangular shape to crop
	# define subplot
	pyplot.subplot(1, len(faces), i+1)
	pyplot.axis('off')
	# plot face
	pyplot.imshow(img[y1:y2, x1:x2])
	# show the plot
pyplot.show()
 

```
